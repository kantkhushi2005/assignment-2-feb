Q-1 Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a powerful Python data visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Here are five plots that can be created using Seaborn, along with their uses:

1. **Scatter Plot**:
   - **Usage**: Used to plot individual data points and observe the relationship between two continuous variables.
   - **Example**: 
     ```python
     import seaborn as sns
     sns.scatterplot(x='total_bill', y='tip', data=tips)
     ```
   - **Use Case**: Analyzing the correlation between the total bill and the tip amount in a restaurant dataset.

2. **Box Plot**:
   - **Usage**: Used to display the distribution of a dataset based on a five-number summary (minimum, first quartile, median, third quartile, and maximum), and to identify outliers.
   - **Example**: 
     ```python
     sns.boxplot(x='day', y='total_bill', data=tips)
     ```
   - **Use Case**: Comparing the distribution of total bills across different days of the week in a restaurant dataset.

3. **Histogram**:
   - **Usage**: Used to plot the distribution of a single continuous variable, showing the frequency of different value ranges.
   - **Example**: 
     ```python
     sns.histplot(tips['total_bill'], bins=10)
     ```
   - **Use Case**: Understanding the distribution of total bills in a restaurant dataset.

4. **Heatmap**:
   - **Usage**: Used to visualize data in a matrix format, where values are represented by different colors. It is useful for displaying the relationship between two variables in a tabular format.
   - **Example**: 
     ```python
     sns.heatmap(correlation_matrix, annot=True)
     ```
   - **Use Case**: Displaying the correlation matrix of a dataset to understand the relationship between different variables.

5. **Pair Plot**:
   - **Usage**: Used to plot pairwise relationships in a dataset, showing scatter plots for each pair of variables and histograms or KDE plots for the distribution of individual variables.
   - **Example**: 
     ```python
     sns.pairplot(tips)
     ```
   - **Use Case**: Exploring the relationships between multiple variables in a dataset, such as total bill, tip amount, and size of the party in a restaurant dataset.

These plots provide a range of options for visualizing different types of data and relationships between variables, making Seaborn a versatile tool for data analysis and exploration.

Q-2 Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri = sns.load_dataset("fmri")

# Create a line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")

# Customize the plot
plt.title('fMRI Signal over Time')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.legend(title='Event and Region')
plt.show()
````

Q-3 Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic = sns.load_dataset("titanic")

# Create the first box plot for age
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x="pclass", y="age")
plt.title('Age Distribution by Pclass')

# Create the second box plot for fare
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x="pclass", y="fare")
plt.title('Fare Distribution by Pclass')

# Show the plots
plt.tight_layout()
plt.show()
````

Q-4 Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds = sns.load_dataset("diamonds")

# Create a histogram for the 'price' column with 'cut' as the hue
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack', bins=50)

# Customize the plot
plt.title('Price Distribution by Cut')
plt.xlabel('Price')
plt.ylabel('Count')
plt.legend(title='Cut')
plt.show()
````

Q-5 Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset
iris = sns.load_dataset("iris")

# Create a pair plot
sns.pairplot(iris, hue="species")

# Customize the plot
plt.title('Pair Plot of Iris Dataset')
plt.show()
````

Q-6 Use the "flights" dataset from seaborn to plot a heatmap.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights = sns.load_dataset("flights")

# Pivot the dataset
flights_pivot = flights.pivot("month", "year", "passengers")

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt="d")

# Customize the plot
plt.title('Number of Passengers per Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()

````

Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose
you have collected data on the amount of time students spend studying for an exam and their final exam
scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.

The Pearson correlation coefficient (r) quantifies the linear relationship between two variables. To calculate it for the amount of time students spend studying and their final exam scores, follow these steps:

1. **Collect Data:** Gather the pairs of values (study time, exam scores).

2. **Calculate Means:** Compute the mean of study times (\(\bar{x}\)) and the mean of exam scores (\(\bar{y}\)).

3. **Compute Differences:** Subtract the means from each individual value to find \( (x_i - \bar{x}) \) and \( (y_i - \bar{y}) \).

4. **Calculate Products:** Multiply these differences for each pair to get \( (x_i - \bar{x})(y_i - \bar{y}) \).

5. **Sum Products:** Sum these products.

6. **Compute Squared Differences:** Sum the squares of the differences for both variables, \( \sum (x_i - \bar{x})^2 \) and \( \sum (y_i - \bar{y})^2 \).

7. **Calculate r:** Use the formula
\[ r = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum (x_i - \bar{x})^2 \cdot \sum (y_i - \bar{y})^2}} \]

### Interpretation:
- \( r \) ranges from -1 to 1.
- \( r = 1 \) indicates a perfect positive linear relationship.
- \( r = -1 \) indicates a perfect negative linear relationship.
- \( r = 0 \) suggests no linear relationship.
- Values close to 1 or -1 indicate a strong linear relationship, while values close to 0 indicate a weak linear relationship.

A high positive \( r \) would suggest that more study time is associated with higher exam scores. Conversely, a high negative \( r \) would suggest that more study time is associated with lower exam scores.


Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables.
Suppose you have collected data on the amount of sleep individuals get each night and their overall job
satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two
variables and interpret the result.

To calculate Spearman's rank correlation coefficient (ρ) for the amount of sleep individuals get each night and their job satisfaction level, follow these steps:

1. **Collect Data:** Gather the pairs of values (amount of sleep, job satisfaction).

2. **Rank Data:** Assign ranks to the values of each variable. If there are ties (i.e., identical values), assign the average rank to each tied value.

3. **Calculate Differences:** Compute the difference (\(d_i\)) between the ranks of each pair of observations.

4. **Square Differences:** Square these differences (\(d_i^2\)).

5. **Sum Squared Differences:** Sum these squared differences (\(\sum d_i^2\)).

6. **Calculate ρ:** Use the formula
\[ \rho = 1 - \frac{6 \sum d_i^2}{n(n^2 - 1)} \]
where \( n \) is the number of pairs.

### Example:
Let's illustrate with a small dataset:
| Sleep (hours) | Job Satisfaction (1-10) |
|---------------|--------------------------|
| 8             | 9                        |
| 6             | 6                        |
| 7             | 7                        |
| 5             | 5                        |
| 9             | 10                       |

1. **Rank Data:**
   - Sleep: 5 (1), 6 (3), 7 (2), 8 (4), 9 (5)
   - Satisfaction: 9 (4), 6 (2), 7 (3), 5 (1), 10 (5)

2. **Calculate Differences:**
   - Differences (d_i): -3, 1, -1, 3, 0

3. **Square Differences:**
   - Squared Differences (d_i^2): 9, 1, 1, 9, 0

4. **Sum Squared Differences:**
   - \(\sum d_i^2 = 20\)

5. **Calculate ρ:**
\[ \rho = 1 - \frac{6 \times 20}{5(5^2 - 1)} = 1 - \frac{120}{120} = 0 \]

### Interpretation:
- \( \rho \) ranges from -1 to 1.
- \( \rho = 1 \) indicates a perfect positive monotonic relationship.
- \( \rho = -1 \) indicates a perfect negative monotonic relationship.
- \( \rho = 0 \) suggests no monotonic relationship.
- Values close to 1 or -1 indicate a strong monotonic relationship, while values close to 0 indicate a weak monotonic relationship.

A high positive \( \rho \) would suggest that more sleep is associated with higher job satisfaction. Conversely, a high negative \( \rho \) would suggest that more sleep is associated with lower job satisfaction. In this example, \( \rho = 0 \) indicates no monotonic relationship between sleep and job satisfaction for the small sample provided.

Q6. A company is interested in examining the relationship between the number of sales calls made per day
and the number of sales made per week. The company collected data on both variables from a sample of
30 sales representatives. Calculate the Pearson correlation coefficient between these two variables.

To calculate the Pearson correlation coefficient between the number of sales calls made per day and the number of sales made per week for 30 sales representatives, follow these steps:

### Step-by-Step Calculation:

1. **Collect Data:** Assume we have data for 30 sales representatives.

2. **Calculate Means:**
   - Mean of sales calls per day (\(\bar{x}\)).
   - Mean of sales made per week (\(\bar{y}\)).

3. **Compute Differences:**
   - Calculate the difference from the mean for each value: \((x_i - \bar{x})\) and \((y_i - \bar{y})\).

4. **Calculate Products:**
   - Compute the product of these differences for each pair: \((x_i - \bar{x})(y_i - \bar{y})\).

5. **Sum Products:**
   - Sum these products.

6. **Compute Squared Differences:**
   - Sum the squared differences for both variables: \(\sum (x_i - \bar{x})^2\) and \(\sum (y_i - \bar{y})^2\).

7. **Calculate Pearson's r:**
   \[
   r = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum (x_i - \bar{x})^2 \cdot \sum (y_i - \bar{y})^2}}
   \]

### Example Dataset:
Let's assume we have the following dataset for 5 representatives (you'll extend this to 30):

| Sales Calls per Day (x) | Sales per Week (y) |
|-------------------------|---------------------|
| 10                      | 15                  |
| 20                      | 30                  |
| 30                      | 45                  |
| 40                      | 50                  |
| 50                      | 60                  |

### Calculation:

1. **Calculate Means:**
   \[
   \bar{x} = \frac{10 + 20 + 30 + 40 + 50}{5} = \frac{150}{5} = 30
   \]
   \[
   \bar{y} = \frac{15 + 30 + 45 + 50 + 60}{5} = \frac{200}{5} = 40
   \]

2. **Compute Differences:**
   \[
   \begin{array}{ccccc}
   x_i & y_i & (x_i - \bar{x}) & (y_i - \bar{y}) & (x_i - \bar{x})(y_i - \bar{y}) \\
   10 & 15 & 10 - 30 = -20 & 15 - 40 = -25 & (-20) \times (-25) = 500 \\
   20 & 30 & 20 - 30 = -10 & 30 - 40 = -10 & (-10) \times (-10) = 100 \\
   30 & 45 & 30 - 30 = 0 & 45 - 40 = 5 & 0 \times 5 = 0 \\
   40 & 50 & 40 - 30 = 10 & 50 - 40 = 10 & 10 \times 10 = 100 \\
   50 & 60 & 50 - 30 = 20 & 60 - 40 = 20 & 20 \times 20 = 400 \\
   \end{array}
   \]

3. **Sum Products:**
   \[
   \sum (x_i - \bar{x})(y_i - \bar{y}) = 500 + 100 + 0 + 100 + 400 = 1100
   \]

4. **Compute Squared Differences:**
   \[
   \sum (x_i - \bar{x})^2 = (-20)^2 + (-10)^2 + 0^2 + 10^2 + 20^2 = 400 + 100 + 0 + 100 + 400 = 1000
   \]
   \[
   \sum (y_i - \bar{y})^2 = (-25)^2 + (-10)^2 + 5^2 + 10^2 + 20^2 = 625 + 100 + 25 + 100 + 400 = 1250
   \]

5. **Calculate Pearson's r:**
   \[
   r = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum (x_i - \bar{x})^2 \cdot \sum (y_i - \bar{y})^2}} = \frac{1100}{\sqrt{1000 \cdot 1250}} = \frac{1100}{\sqrt{1250000}} = \frac{1100}{1118.03} \approx 0.984
   \]

### Interpretation:
The Pearson correlation coefficient \( r \approx 0.984 \) indicates a very strong positive linear relationship between the number of sales calls made per day and the number of sales made per week in this sample dataset. This means that as the number of sales calls increases, the number of sales made also tends to increase significantly.
